```python
##1
# Function to add two matrices
def add_matrices(matrix1, matrix2):
    result = []
    for i in range(len(matrix1)):
        row = []
        for j in range(len(matrix1[0])):
            row.append(matrix1[i][j] + matrix2[i][j])
        result.append(row)
    return result

# Prompt user for input
rows = int(input("Enter the number of rows: "))
columns = int(input("Enter the number of columns: "))

# Initialize the matrices
matrix1 = []
matrix2 = []

print("Enter the elements of matrix 1:")
for i in range(rows):
    row = []
    for j in range(columns):
        element = int(input("Enter element at position ({},{})".format(i, j)))
        row.append(element)
    matrix1.append(row)

print("Enter the elements of matrix 2:")
for i in range(rows):
    row = []
    for j in range(columns):
        element = int(input("Enter element at position ({},{})".format(i, j)))
        row.append(element)
    matrix2.append(row)

# Add the matrices
result_matrix = add_matrices(matrix1, matrix2)
print("The result of adding the two matrices is:")
for row in result_matrix:
    print(row)
```

```python
##2
# Function to multiply two matrices
def multiply_matrices(matrix1, matrix2):
    result = []
    for i in range(len(matrix1)):
        row = []
        for j in range(len(matrix2[0])):
            element = 0
            for k in range(len(matrix2)):
                element += matrix1[i][k] * matrix2[k][j]
            row.append(element)
        result.append(row)
    return result

# Prompt user for input
rows1 = int(input("Enter the number of rows for matrix 1: "))
columns1 = int(input("Enter the number of columns for matrix 1: "))
rows2 = int(input("Enter the number of rows for matrix 2: "))
columns2 = int(input("Enter the number of columns for matrix 2: "))

# Check if matrix multiplication is possible
if columns1 != rows2:
    print("Matrix multiplication is not possible!")
else:
    # Initialize the matrices
    matrix1 = []
    matrix2 = []

    print("Enter the elements of matrix 1:")
    for i in range(rows1):
        row = []
        for j in range(columns1):
            element = int(input("Enter element at position ({},{})".format(i, j)))
            row.append(element)
        matrix1.append(row)

    print("Enter the elements of matrix 2:")
    for i in range(rows2):
        row = []
        for j in range(columns2):
            element = int(input("Enter element at position ({},{})".format(i, j)))
            row.append(element)
        matrix2.append(row)

    # Multiply the matrices
    result_matrix = multiply_matrices(matrix1, matrix2)
    print("The result of multiplying the two matrices is:")
    for row in result_matrix:
        print(row)
```

```python
##3
# Function to transpose a matrix
def transpose_matrix(matrix):
    transpose = []
    for j in range(len(matrix[0])):
        row = []
        for i in range(len(matrix)):
            row.append(matrix[i][j])
        transpose.append(row)
    return transpose

# Prompt user for input
rows = int(input("Enter the number of rows: "))
columns = int(input("Enter the number of columns: "))

# Initialize the matrix
matrix = []

print("Enter the elements of the matrix:")
for i in range(rows):
    row = []
    for j in range(columns):
        element = int(input("Enter element at position ({},{})".format(i, j)))
        row.append(element)
    matrix.append(row)

# Transpose the matrix
transposed_matrix = transpose_matrix(matrix)
print("The transposed matrix is:")
for row in transposed_matrix:
    print(row)


```

```python
##4
# Function to sort words in alphabetic order
def sort_words(words):
    return sorted(words)

# Prompt user for input
sentence = input("Enter a sentence: ")

# Split the sentence into words
words = sentence.split()

# Sort the words
sorted_words = sort_words(words)

# Display the sorted words
print("The sorted words are:", " ".join(sorted_words))

```

```python
##5
import string

# Function to remove punctuation from a string
def remove_punctuation(text):
    translator = str.maketrans("", "", string.punctuation)
    return text.translate(translator)

# Prompt user for input
text = input("Enter a string: ")

# Remove punctuation from the string
text_without_punctuation = remove_punctuation(text)

# Display the string without punctuation
print("The string without punctuation is:", text_without_punctuation)
```
